```python
import cv2

from google.colab.patches import cv2_imshow
```

```python
pic = cv2.imread("/content/dog.jpg")
cv2_imshow(pic)
```

```python
gray_scale_pic = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY) #####-----------> gray image
cv2_imshow(gray_scale_pic)
```

```python
inverted_pic = cv2.bitwise_not(gray_scale_pic)
cv2_imshow(inverted_pic)
```

```python
pic_smoothing = cv2.GaussianBlur(inverted_pic, (15, 21), 0)  #### here we have use Gaussian blur
cv2_imshow(pic_smoothing)
```

sketch fucntion to convert


```python
def pencil_sketch(x,y):
   return cv2.divide(x, 255 - y, scale=256)
```

```python
sketch = pencil_sketch(gray_scale_pic, pic_smoothing)
cv2_imshow(sketch)

cv2_imshow(pic)
```
